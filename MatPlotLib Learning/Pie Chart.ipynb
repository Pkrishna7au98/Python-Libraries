## A chart with so many inputs

```python
clr = ['c','r','b','k','y']
ede=[0,0.2,0,0.2,0]
tp = {'fontsize':15}
wp = {'linewidth':4,'width':1,'edgecolor':"g"}

plt.pie(class1,
    explode=ede,
    labels=classes,
    colors=clr,
    autopct='%0.3f%%',
    pctdistance=0.6,
    shadow=True,
    labeldistance=1.4,
    startangle=280,
    radius=1.2,
    counterclock=False,
    wedgeprops=wp,
    textprops=tp,
    center=(0, 0),
    frame=True,
    rotatelabels=True,)
plt.show()
```

```python
plt.figure(figsize=(3,2))
clr = ['c','r','b','k','y']
ede=[0,0.2,0,0.2,0]
tp = {'fontsize':15}
wp = {'linewidth':4,'width':1,'edgecolor':"g"}

plt.pie(class1,
    explode=ede,
    labels=classes,
    colors=clr,
    autopct='%0.3f%%',
    pctdistance=0.6,
    shadow=True,
    labeldistance=1.4,
    startangle=280,
    radius=1.2,
    counterclock=False,
    wedgeprops=wp,
    textprops=tp,
    center=(0, 0),
    frame=True,
    rotatelabels=True,)
plt.legend()
plt.show()
```
