```python
import matplotlib.pyplot as plt
```

```python
plt.subplot(2,2,1)       # (2,2,1) represents 2 rows , 2 columns and the first of these.
plt.pie([1])             # so this way max 4 plots can be generated through this
                         # (a,b,n) n <= a*b (a rows and b columns)
    
#------------------------    Plot 2
plt.subplot(2,2,2)       
plt.pie([1,2])
#------------------------    Plot 3
plt.subplot(2,2,3)
plt.pie([1,2,3])
#------------------------   Plot 4
plt.subplot(2,2,4)
plt.pie([1,2,3,4]) 
#------------------------  
plt.show()
```
