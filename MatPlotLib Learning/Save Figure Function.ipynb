```python
import matplotlib.pyplot as plt
```

```python
plt.pie([14,23,34])
plt.savefig('pIECHART.jpg')
plt.show()
```

```python
plt.figure(figsize = (16,9))
plt.pie([14,23,34,32,34,32,53,23,12])
plt.savefig("Pie_Chart",
            dpi=1000,
            facecolor='w',
            edgecolor='k',
          orientation='portrait',
            format=None,
          transparent=False,
            bbox_inches=None,
            pad_inches=0.1,
            metadata=None)
```
