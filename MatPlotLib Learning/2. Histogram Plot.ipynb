```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
age1 = np.random.randint(18,50,(100))
age2 = np.random.randint(15,40,(100))
```

```python
print(age1)
print(age2)
```

To use HISTOGRAM plot, we use HIST

```python
plt.hist(age1)
plt.show()
```

```python
plt.hist(age1)
plt.title('ML students Age')
plt.xlabel('student age catagory')
plt.ylabel('No. students age')
plt.show()
```

```python
plt.hist(age2)
plt.title('ML students Age')
plt.xlabel('student age catagory')
plt.ylabel('No. students age')
plt.show()
```

## bins , rwidth

```python
bins = [15,20,25,30,35,40,45] #defining catagory of age groups
plt.hist(age1,bins)
plt.title('ML students Age')
plt.xlabel('student age catagory')
plt.ylabel('No. students age')
plt.show()
```

```python
bins = [15,20,25,30,35,40,45] 
plt.hist(age1,bins,rwidth=0.6)
plt.title('ML students Age')
plt.xlabel('student age catagory')
plt.ylabel('No. students age')
plt.show()
```

## histtype , align , orientation, color , label

```python
bins = [15,20,25,30,35,40,45] 
plt.hist(age1,bins,rwidth=0.6 , histtype = 'step')
plt.title('ML students Age')
plt.xlabel('student age catagory')
plt.ylabel('No. students age')
plt.show()
```

```python
bins = [15,20,25,30,35,40,45] 
plt.hist(age1,bins,rwidth=0.6 , histtype = 'bar')
plt.title('ML students Age')
plt.xlabel('student age catagory')
plt.ylabel('No. students age')
plt.show()
```

```python
bins = [15,20,25,30,35,40,45] 
plt.hist(age1,bins,rwidth=0.6 , histtype = 'bar',align = 'right')
plt.title('ML students Age')
plt.xlabel('student age catagory')
plt.ylabel('No. students age')
plt.show()
```

```python
bins = [15,20,25,30,35,40,45] 
plt.hist(age1,bins,rwidth=0.6 ,orientation = 'horizontal')
plt.title('ML students Age')
plt.xlabel('student age catagory')
plt.ylabel('No. students age')
plt.show()
```

```python
bins = [15,20,25,30,35,40,45] 
plt.hist(age1,bins,rwidth=0.6 , histtype = 'bar' , color='y')
plt.title('ML students Age')
plt.xlabel('student age catagory')
plt.ylabel('No. students age')
plt.show()
```

```python
bins = [15,20,25,30,35,40,45] 
plt.hist(age1,bins,rwidth=0.6 , histtype = 'bar', label = 'ML Studnets')
plt.title('ML students Age')
plt.xlabel('student age catagory')
plt.ylabel('No. students age')
plt.legend()
plt.show()
```

## To print multiple plots Together

```python
bins = [15,20,25,30,35,40,45] 
plt.hist(age1,bins,rwidth=0.6 , histtype = 'bar', label = 'ML Studnets',orientation = 'horizontal')
plt.hist(age2,bins,rwidth=0.6 , histtype = 'bar', label = 'Py Studnets')

plt.title('ML students Age')
plt.xlabel('student age catagory')
plt.ylabel('No. students age')
plt.legend()
plt.show()
```

```python
bins = [15,20,25,30,35,40,45] 
plt.hist([age1,age2],bins,rwidth=0.6 , label = ['ML Studnets','Py Students'])
#plt.hist(age2,bins,rwidth=0.6 , histtype = 'bar', label = 'Py Studnets')

plt.title('ML students Age')
plt.xlabel('student age catagory')
plt.ylabel('No. students age')
plt.legend()
plt.show()
```

```python
bins = [15,20,25,30,35,40,45] 

plt.figure(figsize = (16,9))   # To increase the size of the Histogram

plt.hist([age1,age2],bins,rwidth=0.6 , label = ['ML Studnets','Py Students'])
#plt.hist(age2,bins,rwidth=0.6 , histtype = 'bar', label = 'Py Studnets')

plt.title('ML students Age')
plt.xlabel('student age catagory')
plt.ylabel('No. students age')
plt.legend()
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]

from matplotlib import style
style.use('ggplot')

plt.hist([age1,age2],bins,rwidth=0.6 , label = ['ML Studnets','Py Students'])
#plt.hist(age2,bins,rwidth=0.6 , histtype = 'bar', label = 'Py Studnets')

plt.title('ML students Age')
plt.xlabel('student age catagory')
plt.ylabel('No. students age')
plt.legend()
plt.show()
```
