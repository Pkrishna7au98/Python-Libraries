## Multiple Charts display

```python
plt.bar(classes,class1, edgecolor='b', 
        label="Class 1 Students")
plt.bar(classes,class2, edgecolor='y', 
        label="Class 2 Students")

plt.title('Programming students chart')
plt.xlabel('Classes')
plt.ylabel('No. of Students')
plt.legend()
```

```python
classes_index = np.arange(len(classes))

width = 0.2
plt.bar(classes_index,class1, edgecolor='b', 
        label="Class 1 Students")
plt.bar(classes_index + width,class2, edgecolor='y', 
        label="Class 2 Students")
plt.xticks(classes_index + width, classes)

plt.title('Programming students chart')
plt.xlabel('Classes')
plt.ylabel('No. of Students')
plt.legend()
```

```python
classes_index = np.arange(len(classes))

width = 0.2
plt.bar(classes_index,class1, edgecolor='b', 
        label="Class 1 Students")
plt.bar(classes_index + width,class2, edgecolor='y', 
        label="Class 2 Students")
plt.xticks(classes_index + width, classes)

plt.title('Programming students chart')
plt.ylabel('No. of Students')
plt.legend()
```

```python
classes_index = np.arange(len(classes))

width = 0.2
plt.bar(classes_index,class1, edgecolor='b', 
        label="Class 1 Students")
plt.bar(classes_index + width,class2, edgecolor='y', 
        label="Class 2 Students")
plt.xticks(classes_index + width, classes , rotation = 45)

plt.title('Programming students chart')
plt.ylabel('No. of Students')
plt.legend()
```

```python
plt.figure(figsize=(16,9))

classes_index = np.arange(len(classes))

width = 0.2
plt.bar(classes_index,class1, edgecolor='b',  width = 0.2,
        label="Class 1 Students")
plt.bar(classes_index + width,class2, edgecolor='y', width = 0.2,
        label="Class 2 Students")
plt.bar(classes_index + width + width,class3, edgecolor='r',  width = 0.2,
        label="Class 3 Students")
plt.xticks(classes_index + width, classes , rotation = 45,fontsize = 25)

plt.title('Programming students chart')
plt.ylabel('No. of Students')
plt.legend()
```
