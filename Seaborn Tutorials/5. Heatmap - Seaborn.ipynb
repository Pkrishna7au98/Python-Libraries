```python
annot_arr = np.arange(2,202,2).reshape(10,10)
annot_arr
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(data, vmin = 0 , vmax=100 , cmap ="hot" ,annot=annot_arr , fmt="d")
```

```python
plt.figure(figsize=(8,4.5))
annot_kws = {'color':'b','alpha':0.6}
sns.heatmap(data, vmin = 0 , vmax=100 , cmap ="hot" ,annot=True , annot_kws = annot_kws)
```

```python
plt.figure(figsize=(8,4.5))
annot_kws = {'color':'b','alpha':0.6}
sns.heatmap(data, vmin = 0 , vmax=100 , cmap ="hot" ,annot=True , annot_kws = annot_kws , linewidth= 4)
```

```python
plt.figure(figsize=(8,4.5))
annot_kws = {'color':'b','alpha':0.6}
sns.heatmap(data, vmin = 0 , vmax=100 , cmap ="hot" ,annot=True , annot_kws = annot_kws , linewidth= 4 , linecolor ="y")
```

```python
plt.figure(figsize=(8,4.5))
annot_kws = {'color':'b','alpha':0.6}
sns.heatmap(data, vmin = 0 , vmax=100 , cmap ="hot" ,annot=True , annot_kws = annot_kws , cbar=False)
```

```python
plt.figure(figsize=(8,4.5))
annot_kws = {'color':'b','alpha':0.6}
sns.heatmap(data, vmin = 0 , vmax=100 , cmap ="hot" ,annot=True , annot_kws = annot_kws , cbar=False , xticklabels= True, yticklabels= False)
```

```python
plt.figure(figsize=(16,9))
annot_kws = {'color':'b','alpha':0.6}
cbar_kws = {'orientation':"horizontal", 'shrink':1,'extend':'min','extendfrac':0.1,'drawedges':True}
ax = sns.heatmap(data, vmin = 0 , vmax=100 , cmap ="hot" ,annot=True , annot_kws = annot_kws , cbar_kws = cbar_kws)
ax.set(title="Random understandable plot", xlabel = 'Numbers' , ylabel ='indexes')
sns.set(font_scale = 8)
```

## Correlation Heatmap - Feature Engineering and etc

```python
df2.corr()
```

```python
ax = sns.heatmap(df2.corr(),annot=True , linewidth =3)
ax.tick_params(size=10,color='w',labelsize=10,labelcolor='w')
plt.title('Heatmap of titanic correlations',fontsize = 15)
plt.show()
```

## New Important Program

```python
from sklearn.datasets import load_breast_cancer
cancer_dataset = load_breast_cancer()
cancer_dataset
```

```python
cancer_df  = pd.DataFrame(np.c_[cancer_dataset['data'],cancer_dataset['target']],
            columns = np.append(cancer_dataset['feature_names'],['target']))
cancer_df
```

```python
plt.figure(figsize=(16,9))
ax = sns.heatmap(cancer_df.corr(),annot=True,linewidth=3 )
ax.tick_params(size=10,color='w',labelsize=10,labelcolor='b')
plt.title('Heatmap of breast cancer data',fontsize = 25)
plt.show()
```
