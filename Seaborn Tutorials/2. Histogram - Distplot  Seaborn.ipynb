```python
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
tips_df = sns.load_dataset("tips")
tips_df
```

```python
sns.displot(tips_df["size"])
plt.show()
```

In the above output the line curve is too given  : It is known as KDE : Kernel Density Estimate/or

```python
sns.distplot(tips_df["tip"])
plt.show()
```

```python
sns.distplot(tips_df["tip"] , bins=100)
plt.show()
```

```python
sns.distplot(tips_df["tip"] , bins=100, hist = False)
plt.show()
```

```python
sns.distplot(tips_df["tip"] , bins=100, hist = True , kde = False)
plt.show()
```

```python
sns.distplot(tips_df["tip"] , bins=100, hist = False, rug = True)
plt.show()
```

```python
sns.distplot(tips_df["tip"] , bins=100, hist = False , rug = True , fit = norm)
plt.show()
```

```python
 sns.distplot(tips_df["tip"] , bins=100, hist = True , rug = True , fit = norm , color = "r")
plt.show()
```

```python
 sns.distplot(tips_df["tip"] , bins=100, hist = True , rug = True , fit = norm , color = "r", norm_hist = True, axlabel = "Tips given")
plt.show()
```

```python
sns.distplot(tips_df["tip"] , bins=100, hist = True , rug = True , fit = norm , color = "r", norm_hist = True, axlabel = "Tips given", label = "Plot Info")
plt.legend()
plt.show()
```

```python
sns.histplot(tips_df["tip"] , bins=50  , color = "r", label = "Plot Info", kde =True)
plt.legend()
plt.show()
```

```python
sns.set()
sns.histplot(tips_df["tip"] , bins=50  , color = "r", label = "Plot Info", kde =True)
plt.legend()
plt.show()
```

```python
bins = [1,5,10,15,20,25,30,35,40,45,50,55]
sns.set()
sns.distplot(tips_df["total_bill"] , bins=bins)
plt.xticks(bins)
plt.show()
```

```python
sns.set()
sns.distplot(tips_df["total_bill"] , hist_kws = {'color':'#DC143C'})
```

```python
sns.set()
sns.distplot(tips_df["total_bill"] , hist_kws = {'color':'#DC143C', 'edgecolor':'#aaff00','linewidth':5 , 'linestyle':"--"})
plt.show()
```

```python
plt.figure(figsize=(16,9))
sns.set()
sns.distplot(tips_df["total_bill"] , hist_kws = {'color':'#DC143C', 'edgecolor':'#aaff00','linewidth':5 , 'linestyle':"--",'alpha':0.8})
plt.show()
```

```python
plt.figure(figsize=(16,9))
sns.set()
sns.distplot(tips_df["total_bill"] , hist_kws = {'color':'#DC143C', 'edgecolor':'#aaff00','linewidth':5 , 'linestyle':"--",'alpha':0.8}, kde_kws = {'color':'#8e00ce','linewidth':7 , 'linestyle':"--",'alpha':0.8})
plt.show()
```

```python
plt.figure(figsize=(16,9))
sns.set()
sns.distplot(tips_df["total_bill"] , hist_kws = {'color':'#DC143C', 'edgecolor':'#aaff00','linewidth':5 , 'linestyle':"--",'alpha':0.8},
             kde_kws = {'color':'#8e00ce','linewidth':7 , 'linestyle':"--",'alpha':0.8} ,
             rug=True )
plt.show()
```

```python
plt.figure(figsize=(16,9))
sns.set()
sns.distplot(tips_df["total_bill"] , hist_kws = {'color':'#DC143C', 'edgecolor':'#aaff00','linewidth':5 , 'linestyle':"--",'alpha':0.8},
             kde = True,
             kde_kws = {'color':'#8e00ce','linewidth':7 , 'linestyle':"--",'alpha':0.8} ,
             rug=True,
             rug_kws = {'color':'#8e00ce','linewidth':7 , 'linestyle':"--",'alpha':0.8})
plt.show()
```

```python
plt.figure(figsize=(16,9))
sns.set()
sns.distplot(tips_df["total_bill"] , hist_kws = {'color':'#DC143C', 'edgecolor':'#aaff00','linewidth':5 , 'linestyle':"--",'alpha':0.8},
             kde = False,
             fit = norm,
             fit_kws = {'color':'#8e00ce','linewidth':7 , 'linestyle':"--",'alpha':0.8} ,
             rug=True,
             rug_kws = {'color':'#8e00ce','linewidth':7 , 'linestyle':"--",'alpha':0.8})
plt.show()
```

```python
plt.figure(figsize=(16,9))
sns.set()
sns.distplot(tips_df["total_bill"] , hist_kws = {'color':'#DC143C', 'edgecolor':'#aaff00','linewidth':5 , 'linestyle':"--",'alpha':0.8},
             kde = False,
             fit = norm,
             fit_kws = {'color':'#8e00ce','linewidth':7 , 'linestyle':"--",'alpha':0.8} ,
             rug=True,
             rug_kws = {'color':'#8e00ce','linewidth':7 , 'linestyle':"--",'alpha':0.8},
             label = "TB")
plt.title("Histogram of Restaurant Bills")
plt.xlabel('Total Bill')
plt.show()
```
