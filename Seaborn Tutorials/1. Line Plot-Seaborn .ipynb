```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
days = list(range(1,16))
temperature = [45,56,34,43,45,56,45,76,87,65,34,56,76,76,87]
dataframe = pd.DataFrame({'days':days , 'temperature':temperature})
sns.lineplot(x = 'days',y = 'temperature' , data = dataframe)
plt.show()
```

In seaborn it automatically gives the labels 

To use all the parameters we need to use proper dataframe. So get it from github directly without downloading.  >> github seaborn dataset >> tips.csv

```python
tips_df = sns.load_dataset(name = "tips")
tips_df
```

```python
sns.lineplot(x = "total_bill" , y = "tip" , data = tips_df)
```

```python
sns.lineplot(x = "total_bill" , y = "size" , data = tips_df)
```

```python
sns.lineplot(x = "size" , y = "total_bill" , data = tips_df)
```

```python
sns.lineplot(x = "size" , y = "total_bill" , data = tips_df , hue  ="sex")
```

hue is given = sex catagory :: For each catagory a plot will be generated. Here for MALE and FEMALE , plot is generated.

Here the LEGEND is by default.

```python
sns.lineplot(x = "size" , y = "total_bill" , data = tips_df , hue  ="sex" , style = 'sex')
plt.show()
```

```python
sns.lineplot(x = "size" , y = "total_bill" , data = tips_df , hue  ="sex" , style = 'sex', palette = 'hot')
plt.show()
```

```python
sns.lineplot(x = "size" , y = "total_bill" , data = tips_df , hue  ="sex" , style = 'sex', palette = 'hot' , dashes = False)
plt.show()
```

```python
sns.lineplot(x = "size" , y = "total_bill" , data = tips_df , hue  ="sex" , style = 'sex', palette = 'hot' , dashes = False , markers = ['o', '<'])
plt.show()
```

```python
sns.lineplot(x = "size" , y = "total_bill" , data = tips_df , hue  ="sex" , style = 'sex', palette = 'hot' , dashes = False , markers = ['o', '<'] , legend = "brief")
plt.show()
```

```python
sns.lineplot(x = "size" , y = "total_bill" , data = tips_df , hue  ="sex" , style = 'sex', palette = 'hot' , dashes = False , markers = ['o', '<'] , legend = False)
plt.show()
```

```python
sns.lineplot(x = "size" , y = "total_bill" , data = tips_df , hue  ="sex" , style = 'sex', palette = 'hot' , dashes = False , markers = ['o', '<'] , legend = "full")
plt.title("Line Plot")
plt.xlabel("size", fontsize = 15)
plt.ylabel("Total Bill", fontsize = 15)
plt.show()
```

```python
sns.set(style = 'darkgrid')
sns.lineplot(x = "size" , y = "total_bill" , data = tips_df , hue  ="day" , style = 'day', palette = 'hot' , dashes = False , markers = ['o', '<' , '*' , '>'] , legend = "full")
plt.title("Line Plot")
plt.xlabel("size", fontsize = 15)
plt.ylabel("Total Bill", fontsize = 15)
plt.show()
```

```python
sns.set(style = 'darkgrid')
ho = ('Thur','Sun','Fri','Sat')
sns.lineplot(x = "size" , y = "total_bill" , data = tips_df , hue  ="day" , style = 'day', palette = 'hot' , dashes = False , markers = ['o', '<' , '*' , '>'] , legend = "full", hue_order = ho )
plt.title("Line Plot")
plt.xlabel("size", fontsize = 15)
plt.ylabel("Total Bill", fontsize = 15)
plt.show()
```

```python
sns.set(style = 'darkgrid')
ho = ('Thur','Sun','Fri','Sat')
sns.lineplot(x = "size" , y = "total_bill" , data = tips_df , hue  ="day" , style = 'day', palette = 'hot' , dashes = False , markers = ['o', '<' , '*' , '>'] , legend = "full", hue_order = ho ,sort = False )
plt.title("Line Plot")
plt.xlabel("size", fontsize = 15)
plt.ylabel("Total Bill", fontsize = 15)
plt.show()
```
