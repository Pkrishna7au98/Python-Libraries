```python
x_sin = np.arange(0,3*np.pi,0.2)
print(x_sin)
```

```python
y_sin = np.sin(x_sin)
print(y_sin)
```

```python
pplt.plot(x_sin,y_sin)
pplt.show()
```

```python
y_cos = np.cos(x_sin)
pplt.plot(x_sin,y_cos)
pplt.show()
```

```python
y_tan = np.tan(x_sin)
pplt.plot(x_sin,y_tan)
pplt.show()
```
